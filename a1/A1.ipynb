```python
import numpy as np
import matplotlib.pyplot as plt
```

# Question 1 Code

```python
results = []
phi = np.float64((np.sqrt(5) - 1) / 2)
results.append(phi**0)
results.append(phi**1)

for i in range(2, 101):
    new_phi = results[-2] - results[-1] # phi**(i+2) = phi**(i) - phi**(i+1)
    results.append(new_phi)

print(results)
```

```python
plt.plot(results)
plt.show()
```

```python
log_results = [np.log(np.abs(x)) for x in results]
plt.plot(log_results)
plt.show()
```

```python
def generate_series(x0: np.float64, x1: np.float64, n: int) -> list:
    results = []
    results.append(x0)
    results.append(x1)
    for i in range(2, n):
        new_x = results[-2] - results[-1]
        results.append(new_x)
    return results
```

```python
series = generate_series(np.float64(1), np.float64(phi), 100)
series
```

```python
log_series = [np.log(np.abs(x)) for x in series]
plt.plot(log_series)
plt.show()
```
